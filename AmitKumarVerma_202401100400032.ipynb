```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Import different libraries for statistical calculations
```

```python
# Load the dataset (Make sure the CSV file is in the same directory)
data = pd.read_csv("Employee_Salary_Dataset.csv")
```

```python
# Basic salary statistics (Gives us an idea of the salary distribution)
print("\nSalary Statistics:")
print(data['Salary'].describe())  # Provides summary statistics for salary data
```

```python
# Checking the basic info about the dataset
print(data.info())
print(data.head())
```

```python
#  1. Visualizing Salary Distribution (Histogram + KDE for smoothness)
plt.figure(figsize=(8, 5))
sns.histplot(data['Salary'], bins=30, kde=True, color='blue')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
# Checking for missing values
print("Missing Values:\n", data.isnull().sum())
```

```python
# Printing all th ecoloumns in the data
print(data.columns)
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Experience_Years', y='Salary', data=data, palette='coolwarm')
plt.xticks(rotation=45)
plt.title("Salary by Experience_Years")
plt.xlabel("Experience_Years")
plt.ylabel("Salary")
plt.show()
```

```python
#   Making a correlation Matrix for the salary and the Experience of the employees
correlation_matrix = data[['Salary', 'Experience_Years']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Comparing the Employees Salaries on the basis of their experience
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Experience_Years', y='Salary', data=data, hue='Age', palette='deep')
plt.title("Salary vs. Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend(title="Age", bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Using PairPlot
print(data.info())
sns.pairplot(data, hue='Gender')
plt.show()
```

```python
# Comparing the salary of the emplyees on the basis of their ages
plt.figure(figsize=(12, 6))
sns.boxplot(x='Age', y='Salary', data=data, palette='coolwarm')
plt.xticks(rotation=45)
plt.title("Salary by Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
